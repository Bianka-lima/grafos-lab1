```python
### CÉLULA (7)

B = nx.Graph()

# Adiciona os vértices de R em B
B.add_nodes_from(R.nodes(data=True))
for p in B.nodes:
  B.nodes[p]["type"] = "pessoa"

# Cria os vértices para cada cidade
for i, cidade in enumerate(lista_cidades):
    node_id = f"c{i}"
    B.add_node(node_id, nome=cidade, type="cidade")

# Cria as arestas
cidades = [x for x in B.nodes if B.nodes[x]['type']=="cidade"] # Corrected: use string key 'type'
for p in R.nodes:
  for c in cidades:
    if R.nodes[p]["cidade"]==B.nodes[c]["nome"]:
      B.add_edge(p,c)

draw_graph(B, layoutid="arf_layout")
```
